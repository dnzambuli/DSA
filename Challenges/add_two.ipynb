# Solution using regular lists

```python
def addTwoNumbers(l1, l2):
        longest_distance = max(len(l1), len(l2))
        carry = 0
        out = []
        if len(l1)< len(l2):
            diff  = abs(len(l1) - len(l2))
            for _ in range(diff):
                l1.append(0)
        else:
            diff  = abs(len(l1) - len(l2))
            for _ in range(diff):
                l2.append(0)
                
        # print(l1)
        # print(l2)
        
        for i in range(longest_distance):
            sol = l1[i] + l2[i]
            # print(sol)
            if carry > 0:
                sol += carry
            out.append(sol % 10)
            carry = sol // 10
            
        if carry > 0:
            out.append(carry)
            
        return out
            

print(addTwoNumbers(l1 = [9,9,9,9,9,9,9], l2 = [9,9,9,9]))
```

# Using Linked List

```python
# Using Linked Lists

class Node:
    def __init__(self, value):
        self.val = value
        self.next = None

def add_elements(n, l )-> bool:
    for i in n:
        l.next = Node(i)
        l = l.next


l1 = Node(9)

l1_input = [9,9,9,9,9,9]

add_elements(l1_input, l1)

l2 = Node(9)
l2_input = [9,9,9]

add_elements(l2_input, l2)

def ll_values(l):
    while l:
        print(l.val, end = "--->")
        l = l.next


ll_values(l1)


        
```

## Add two Numbers in a Linked List

```python
def addTwoNumbers_ll(l1, l2) -> Node:
    '''
    idea:
        1. have a zero node (identifies the add 2 numbers output)
        2. point the current sum to the initial zero node (helps create a head node for the new ll)
        3. create a carry forward variable
        4. check for numbers to add or a value to carry 
            5. add the numbers and only keep the sum % 10
            6. update carry 
            7. Move to the next element in both listnodes

    '''
    out = Node(0)
    current = out
    carry = 0

    while l1 or l2 or carry:
        a = l1.val if l1 else 0
        b = l2.val if l2 else 0

        total = a + b + carry
        carry = total // 10
        current.next = Node(total % 10)

        current = current.next
        if l1:
            l1 = l1.next 
        if l2:
            l2 = l2.next

    return out.next

c = addTwoNumbers_ll(l1, l2)

ll_values(c)
```
