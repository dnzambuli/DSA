# Big O O(N)

```python
def print_items(n):
    for i in range(n):
        print(i)

print_items(10)
```

## A Representation of O(n)

```python
import matplotlib.pyplot as plt

data = [0, 1, 2, 3, 4, 5]
plt.plot(data,
         data)

plt.xlabel('Operations Taken')
plt.ylabel('Tasks Completed')
plt.title('O(n) graph')

plt.show()
```

> The graph of O(n) is always linear
>

## Drop Constants Big O


```python
def print_items(n):
    for i in range(n):
        print(i)
    for j in range(n):
        print(j)

print_items(5)
```

### Explanation

The code above runs n + n times `n = 5 in the above case`

the notation is writen as

> the code run O(n + n ) times
>
> or O(2n)
>

But to simplify the notation all we need to do is drop the constant `2`

$\therefore$ the code run **O(n)**

---

O(n) grows directly with the input size. For example, if we loop through a list of ten items, that's ten steps. Double the list to twenty items, and it's twenty steps. It's predictable and super manageable

---

## O($n^2$)

```python
def print_items(n):
    for i in range(n):
        for j in range(n):
            print(i, j)

print_items(5)
```

### What Happened

there are 25 steps taken to solve 5 tasks

$$
O(5 * 5)
$$

$$
O(n^2)
$$

### Representation

```python
import numpy as np
n_sqrd = np.arange(0, 10)
plt.plot(n_sqrd, n_sqrd**2)
plt.plot(n_sqrd, n_sqrd)
plt.xlabel('Operations Taken')
plt.ylabel('Tasks Completed')
plt.title('O(n) graph')

plt.show()
```

---

Imagine, instead of just walking up a flight of stairs, youre checking each step for something, and then for every step, you go back and check every previous step again. So for a house with three steps, you'd make nine checksthree times three. For thirty steps, you'd check nine hundred timesthirty times thirty! That's the steep curve of O(n) complexity. It multiplies the work because everything interacts with everything else.

---

> O(n) is the steady climber
>
> O($n^2) is always out of breathe because of the number of nested operations taken for each increasing step
>
>

---

## O(1)


```python
def add_items(n):
     return n + n

add_items(5)
```

---

O(1) constant time operations are the most efficient.

No mater the size of the task the program only does the given operation(s)

---

### Representation

```python
n_sqrd = np.arange(1, 10)
plt.plot(n_sqrd, n_sqrd)
plt.plot(n_sqrd, n_sqrd**2)
plt.plot(n_sqrd, np.ones_like(n_sqrd))
plt.xlabel('Operations Taken')
plt.ylabel('Tasks Completed')
plt.title('O(n) graph')

plt.show()
```

## O(log(n))

given a sorted collection of numbers

> 1 2 3 4 5 6 7 8
>

The target is to find a given number `1`

### Steps to find the solution

> 1. `1 2 3 4` and `5 6 7 8` because one is not in the second list. Dispose the second list and divide the first list in the middle
>
> 2. `1 2` and `3 4` because one is not in the second list. Dispose the second list and divide the first list in the middle
>
> 3. `1` and `2`. We have found one
>

**Notice**
To find a number form a group of 8 numbers we only needed 3 steps

$$
2^3 = 8
$$

$$
log_{2} 8 = 3
$$

> How many times can you take an item divide it into two parts to get to one item
>


```python
plt.plot(n_sqrd, n_sqrd, label = "O(n)")
plt.plot(n_sqrd, n_sqrd**2, label = "O(n^2)")
plt.plot(n_sqrd, np.ones_like(n_sqrd), label = "O(1)")
plt.plot(n_sqrd, np.log2(n_sqrd), label = "O(log n)")
plt.xlabel('Operations Taken')
plt.ylabel('Tasks Completed')
plt.title('O(n) graph')
plt.legend(loc = 'upper right')

plt.show()
```

```python
plt.plot(n_sqrd, n_sqrd, label = "O(n)")
plt.plot(n_sqrd, n_sqrd**2, label = "O(n^2)")
plt.plot(n_sqrd, np.ones_like(n_sqrd), label = "O(1)")
plt.plot(n_sqrd, np.log2(n_sqrd), label = "O(log n)")
plt.plot(n_sqrd, np.power(np.log2(n_sqrd), 4), label = "O(nlog n)")
plt.xlabel('Operations Taken')
plt.ylabel('Tasks Completed')
plt.title('O(n) graph')
plt.legend(loc = 'upper right')

plt.show()
```
