# iterative approach

````python
# walking 100 steps
def walk(n):
    for i in range(1, n+1):
        print(f"You have taken step: {i}")

walk(100)
````

#  Using recursion

The idea is that after walking `1 step` you have $n-1$ steps left to walk so to walk

> 1. call walk(n)
>
> 2. call walk(n -1) within the function

````python
def walk(steps):
    # base case: where to stop calling the function
    print(f"You have taken step: {steps}")

    walk(steps-1)

walk(100)
````

# Recursion dept exceeded 

Occurs because there is no end point for the function

> **Notice**
>
> the steps keep counting after 0 steps 
>

`Zero` is a base case because at 0 steps we are standing not walking

## Set the base case

````python
def walk(steps):
    if steps == 0:
        return # stops a function call
    print(f"You have taken step: {steps}")

    walk(steps-1)

walk(100)
````

# Why it looks like we are walking backward 

## Call Stacking 

every function call is implemented using a stack where:

1. The output of **the first time** a function is called **is last**
2. The output of **the last time** a function is called **is first**

## Back to code 

```py

        return

    walk(steps-1)
```

1. the print is called before recalling walk(n-1)

2. steps starts at 100, 99, 98, ...

what happens when switched to

````python
def walk(steps):
    if steps == 0:
        return

    walk(steps-1)

    print(f"You have taken step: {steps}")

walk(10)
````

# Iterative Factorial

````python
import time 

def factorial(n):
    if n <= 1:
        return 1
    result = 1
    for i in range(1, n +1):
        result *= i 
    return result
    
def factorial_recursion(n):
    if n == 1:
        return 1
    else:
        return n * factorial_recursion(n-1)
    

start_time = time.time()
print(factorial(10))
print(f"----{time.time()-start_time} seconds ----")

start_time = time.time()
print(factorial_recursion(10))
print(f"----{time.time()-start_time} seconds ----")
````
